```python
# Create a Scatter Plot Using the MATLAB Method
# Two methods plt.plot() and plt.scatter()
```

```python
# Creat a scatter plot with with plt.plot() background info.
# If we were only to use the plt.plot() function with our x-axis and y-axis, it would create a line plot and that's not what we're looking for
# When we use the plt.plot() function to create a scatter plot, we need to add a lowercase "o" as a parameter inside the parentheses. This switches the plot from a line chart to a scatter plot.
```

```python
# Here's the data
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumul...
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.22, 24.12, 41.23]
```

```python
# create a scatter plot using the plt.scatter() function
plt.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the MATLAB Method
# Bubble charts are useful when presenting nancial, population, and weather data because you can add a third and fourth factor to convey more info.
# By changing the "dot" into a "bubble," we are adding a third factor: size
# If there is more than one dataset that uses the same axes, we can change the color of each marker for each dataset, which will add a fourth factor: color.
```

```python
# create a bubble chart by making a simple modication to the existing scatter plot code.
# To change the size of our marker on the graph, we have to use the s parameter
# The s parameter must be equal to a feature of our data
# We will make the size of each marker equal to the data oating-point decimal values of the y-axis by using s=y_axis
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# To adjust the size by multiplying the data in the y-axis by any value
y_axis_larger = []
for data in y_axis:
 y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)

# Alternatively we could've used:
# new_list = [expression for item in list if conditional]
# plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Import dependencies. New dependency - statistics module.
# Statistics module allows us to add y-error bars to line and bar charts
# We can also calculate the standard deviation of the fare amount
import matplotlib.pyplot as plt
import statistics
```

```python
# Here's the data
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumul...
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.22, 24.12, 41.23]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev

#Next, add the x_axis , y_axis , and yerr=stdev to the errorbar function:
plt.errorbar(x_axis, y_axis, yerr=stdev)

# We can make the error bars look more professional by adding a "cap" to the error bars that's equal to a scalar value. To do this, we use the capsize= parameter. 
# Let's add capsize=3 to the error bars. Add the following code:
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# add error bars and a capsize using the object-oriented approach. Add the following to a new cell.
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Add Error Bars to a Bar Chart
# To add error bars and caps to a bar chart using the MATLAB approach, we can use the yerr= and the capsize= parameters with the plt.bar() function
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Now let's create the same graph using the object-oriented approach.
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```
